```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('../glfharris.mplstyle')
```

```python
def sig(x):
    return 1/(1 + np.exp(- 0.2 * (x - 25)))

x = np.linspace(0, 50, 2500)
p = sig(x)

plt.xlim([0,50])
plt.ylim([0,1])

plt.xlabel("Partial Pressure of Oxygen")
plt.ylabel("Oxygen Saturation")

plt.plot(x, p)
plt.show()
```
