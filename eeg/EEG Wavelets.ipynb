```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# Generate synthetic EEG data
fs = 1000  # Sampling frequency
T = 10  # Duration in seconds
t = np.arange(0, T, 1/fs)
frequencies = [2,4,8,16]  # Frequencies in Hz
eeg_data = np.zeros_like(t)
for freq in frequencies:
    eeg_data += np.sin(2 * np.pi * freq * t)

# Apply Morlet wavelet transform
wavelet = 'morl'
scales = np.arange(1, 128)  # Wavelet scales
coefficients, frequencies = pywt.cwt(eeg_data, scales, wavelet)

# Plot EEG data
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, eeg_data)
plt.title('EEG Data')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# Plot wavelet transform
plt.subplot(2, 1, 2)
plt.imshow(np.abs(coefficients), extent=[0, T, 1, 128], cmap='jet', aspect='auto')
plt.colorbar(label='Magnitude')
plt.title('Morlet Wavelet Transform')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.tight_layout()
plt.show()
```
