```python
import numpy as np
import math
import matplotlib.pyplot as plt

plt.style.use('../glfharris.mplstyle')
```

```python
duration = 15
calc_step = 1 # Number of calculations per minute
time = np.linspace(0, duration, duration*calc_step)

# Initialising variables
C0 = 100 # Starting Concentration
k = 0.5 # Rate Constant

# Continuous version described by y = C_0 * e^(-kt)
cont_results = []
for t in time:
    cont_results.append(C0 * math.exp(-k * t))

# Discrete version computed by calculating the change per unit time
disc_results = []
C=C0
for t in time:
    dCdt = -k*C / calc_step
    C += dCdt
    disc_results.append(C)

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.plot(time, disc_results)
ax1.set_xlim([0,15])
ax1.set_ylim([0,100])
ax1.set_title("Discrete")

ax2.plot(time, cont_results)
ax2.set_xlim([0,15])
ax2.set_ylim([0,100])
ax2.set_title("Continuous")

fig.set_figwidth(8)
fig.supxlabel("Time (minutes)")
fig.supylabel("Concentration")
```

```python
calc_step = 60 # Number of calculations per minute
time = np.linspace(0, duration, duration*calc_step)

# Initialising variables
C0 = 0 # Starting Concentration
k = 0.5 # Rate Constant

# Continuous version described by y = C_0 * e^(-kt)
cont_results = []
for t in time:
    # cont_results.append((7 / k) *(1-math.exp(-k * t)))
    cont_results.append((C0 * 5) / (C0 + (5 - C0) * math.exp(k * t)))

# Discrete version computed by calculating the change per unit time
disc_results = []
C=C0
for t in time:
    dCdt = (-k*C + 7) / calc_step 
    C += dCdt
    disc_results.append(C)

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.plot(time, disc_results)
ax1.set_xlim([0,15])
ax1.set_ylim([0,100])
ax1.set_title("Discrete")

ax2.plot(time, cont_results)
ax2.set_xlim([0,15])
ax2.set_ylim([0,100])
ax2.set_title("Continuous")

fig.set_figwidth(8)
fig.supxlabel("Time (minutes)")
fig.supylabel("Concentration")
```
